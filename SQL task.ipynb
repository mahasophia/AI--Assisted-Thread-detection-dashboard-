```python
#SELECT Queries
import sqlite3

# Connect to the database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    Emp_ID INTEGER PRIMARY KEY,
    Emp_Name TEXT,
    Department TEXT,
    Salary REAL
)
""")

# Delete old records (optional)
cursor.execute("DELETE FROM Employee")

# Insert sample records
employees = [
    (101, "John", "HR", 30000),
    (102, "Alice", "Finance", 45000),
    (103, "David", "IT", 55000),
    (104, "Sophia", "Marketing", 40000),
    (105, "James", "IT", 60000)
]

cursor.executemany("INSERT INTO Employee VALUES (?, ?, ?, ?)", employees)
conn.commit()

# 1. Display all records
print("1. All Records")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# 2. Display specific columns
print("\n2. Employee Name and Salary")
cursor.execute("SELECT Emp_Name, Salary FROM Employee")
for row in cursor.fetchall():
    print(row)

# 3. Filter records using WHERE
print("\n3. Employees in IT Department")
cursor.execute("SELECT * FROM Employee WHERE Department='IT'")
for row in cursor.fetchall():
    print(row)

# 4. Use AND condition
print("\n4. IT Department AND Salary > 55000")
cursor.execute("SELECT * FROM Employee WHERE Department='IT' AND Salary > 55000")
for row in cursor.fetchall():
    print(row)

# 5. Use OR condition
print("\n5. HR OR Marketing Department")
cursor.execute("SELECT * FROM Employee WHERE Department='HR' OR Department='Marketing'")
for row in cursor.fetchall():
    print(row)

# 6. Retrieve records using IN
print("\n6. Department IN (HR, Finance)")
cursor.execute("SELECT * FROM Employee WHERE Department IN ('HR','Finance')")
for row in cursor.fetchall():
    print(row)

# 7. Retrieve records using BETWEEN
print("\n7. Salary BETWEEN 35000 AND 60000")
cursor.execute("SELECT * FROM Employee WHERE Salary BETWEEN 35000 AND 60000")
for row in cursor.fetchall():
    print(row)

# 8. Search records using LIKE
print("\n8. Employee Names Starting with 'J'")
cursor.execute("SELECT * FROM Employee WHERE Emp_Name LIKE 'J%'")
for row in cursor.fetchall():
    print(row)

# 9. Sort records using ORDER BY
print("\n9. Employees Sorted by Salary")
cursor.execute("SELECT * FROM Employee ORDER BY Salary DESC")
for row in cursor.fetchall():
    print(row)

# 10. Limit the number of records
print("\n10. First 3 Records")
cursor.execute("SELECT * FROM Employee LIMIT 3")
for row in cursor.fetchall():
    print(row)

# Close connection
conn.close()
```

```python
#SQL Aggregate Functions & Grouping
import sqlite3

# Connect to the database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Create Employee table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Employee (
    Emp_ID INTEGER PRIMARY KEY,
    Emp_Name TEXT,
    Department TEXT,
    Salary REAL
)
""")

# Delete old records (optional)
cursor.execute("DELETE FROM Employee")

# Insert sample records
employees = [
    (101, "John", "HR", 30000),
    (102, "Alice", "Finance", 45000),
    (103, "David", "IT", 55000),
    (104, "Sophia", "Marketing", 40000),
    (105, "James", "IT", 60000),
    (106, "Emma", "Finance", 50000)
]

cursor.executemany("INSERT INTO Employee VALUES (?, ?, ?, ?)", employees)
conn.commit()

# 1. Display unique departments using DISTINCT
print("1. Unique Departments")
cursor.execute("SELECT DISTINCT Department FROM Employee")
for row in cursor.fetchall():
    print(row)

# 2. Count the total number of employees
print("\n2. Total Employees")
cursor.execute("SELECT COUNT(*) FROM Employee")
print(cursor.fetchone()[0])

# 3. Calculate the total salary using SUM()
print("\n3. Total Salary")
cursor.execute("SELECT SUM(Salary) FROM Employee")
print(cursor.fetchone()[0])

# 4. Find the highest salary using MAX()
print("\n4. Highest Salary")
cursor.execute("SELECT MAX(Salary) FROM Employee")
print(cursor.fetchone()[0])

# 5. Find the lowest salary using MIN()
print("\n5. Lowest Salary")
cursor.execute("SELECT MIN(Salary) FROM Employee")
print(cursor.fetchone()[0])

# 6. Group employees by department using GROUP BY
print("\n6. Employees in Each Department")
cursor.execute("""
SELECT Department, COUNT(*) 
FROM Employee
GROUP BY Department
""")
for row in cursor.fetchall():
    print(row)

# 7. Filter grouped results using HAVING
print("\n7. Departments with More Than 1 Employee")
cursor.execute("""
SELECT Department, COUNT(*) 
FROM Employee
GROUP BY Department
HAVING COUNT(*) > 1
""")
for row in cursor.fetchall():
    print(row)

# Close the connection
conn.close()
```

```python
#ALTER TABLE & Table Management
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop old tables if they exist
cursor.execute("DROP TABLE IF EXISTS Employee")
cursor.execute("DROP TABLE IF EXISTS Staff")

# Create Employee table
cursor.execute("""
CREATE TABLE Employee (
    Emp_ID INTEGER PRIMARY KEY,
    Emp_Name TEXT,
    Department TEXT,
    Salary REAL
)
""")

# Insert sample records
employees = [
    (101, "John", "HR", 30000),
    (102, "Alice", "Finance", 45000),
    (103, "David", "IT", 50000)
]

cursor.executemany("INSERT INTO Employee VALUES (?, ?, ?, ?)", employees)
conn.commit()

# 1. Display the existing Employee table
print("1. Existing Employee Table")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# 2. Add a new column
cursor.execute("ALTER TABLE Employee ADD COLUMN Email TEXT")
print("\n2. Email column added.")

# 3. Update values of the new column
cursor.execute("UPDATE Employee SET Email='john@gmail.com' WHERE Emp_ID=101")
cursor.execute("UPDATE Employee SET Email='alice@gmail.com' WHERE Emp_ID=102")
cursor.execute("UPDATE Employee SET Email='david@gmail.com' WHERE Emp_ID=103")
conn.commit()

print("\n3. Email values updated.")

# 4. Rename an existing column
cursor.execute("ALTER TABLE Employee RENAME COLUMN Emp_Name TO Employee_Name")
print("\n4. Column renamed successfully.")

# 5. Rename the table
cursor.execute("ALTER TABLE Employee RENAME TO Staff")
print("\n5. Table renamed to Staff.")

# 6. Display the modified table
print("\n6. Modified Staff Table")
cursor.execute("SELECT * FROM Staff")
for row in cursor.fetchall():
    print(row)

# 7. Drop the table
cursor.execute("DROP TABLE Staff")
print("\n7. Staff table dropped.")

# Commit changes
conn.commit()

# 8. Verify the table has been deleted
cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='Staff'")
result = cursor.fetchone()

if result:
    print("\n8. Table still exists.")
else:
    print("\n8. Verification: Staff table has been deleted.")

# Close the connection
conn.close()
```

```python
#SQL JOIN Operations
import sqlite3

# Connect to the database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop tables if they already exist
cursor.execute("DROP TABLE IF EXISTS Employee")
cursor.execute("DROP TABLE IF EXISTS Department")

# Create Department table
cursor.execute("""
CREATE TABLE Department (
    Dept_ID INTEGER PRIMARY KEY,
    Dept_Name TEXT
)
""")

# Create Employee table
cursor.execute("""
CREATE TABLE Employee (
    Emp_ID INTEGER PRIMARY KEY,
    Emp_Name TEXT,
    Salary REAL,
    Dept_ID INTEGER,
    FOREIGN KEY (Dept_ID) REFERENCES Department(Dept_ID)
)
""")

# Insert records into Department table
departments = [
    (1, "HR"),
    (2, "Finance"),
    (3, "IT")
]

cursor.executemany("INSERT INTO Department VALUES (?, ?)", departments)

# Insert records into Employee table
employees = [
    (101, "John", 30000, 1),
    (102, "Alice", 45000, 2),
    (103, "David", 50000, 3),
    (104, "Sophia", 40000, 2)
]

cursor.executemany("INSERT INTO Employee VALUES (?, ?, ?, ?)", employees)

conn.commit()

# 1. Display Employee table
print("Employee Table")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# 2. Display Department table
print("\nDepartment Table")
cursor.execute("SELECT * FROM Department")
for row in cursor.fetchall():
    print(row)

# 3. INNER JOIN
print("\nINNER JOIN")
cursor.execute("""
SELECT Employee.Emp_Name,
       Department.Dept_Name,
       Employee.Salary
FROM Employee
INNER JOIN Department
ON Employee.Dept_ID = Department.Dept_ID
""")

for row in cursor.fetchall():
    print(row)

# 4. LEFT JOIN
print("\nLEFT JOIN")
cursor.execute("""
SELECT Employee.Emp_Name,
       Department.Dept_Name,
       Employee.Salary
FROM Employee
LEFT JOIN Department
ON Employee.Dept_ID = Department.Dept_ID
""")

for row in cursor.fetchall():
    print(row)

# 5. CROSS JOIN
print("\nCROSS JOIN")
cursor.execute("""
SELECT Employee.Emp_Name,
       Department.Dept_Name
FROM Employee
CROSS JOIN Department
""")

for row in cursor.fetchall():
    print(row)

# Close connection
conn.close()
```

```python
#SQL Subqueries & Views
import sqlite3

# Connect to database
conn = sqlite3.connect("company.db")
cursor = conn.cursor()

# Drop old table and view
cursor.execute("DROP VIEW IF EXISTS Employee_View")
cursor.execute("DROP TABLE IF EXISTS Employee")

# Create Employee table
cursor.execute("""
CREATE TABLE Employee(
    Emp_ID INTEGER PRIMARY KEY,
    Emp_Name TEXT,
    Department TEXT,
    Salary REAL
)
""")

# Insert records
employees = [
    (101,'John','HR',30000),
    (102,'Alice','Finance',45000),
    (103,'David','IT',60000),
    (104,'Sophia','IT',55000),
    (105,'James','Marketing',35000)
]

cursor.executemany("INSERT INTO Employee VALUES(?,?,?,?)",employees)
conn.commit()

# 1. Display Employee table
print("Employee Table")
cursor.execute("SELECT * FROM Employee")
for row in cursor.fetchall():
    print(row)

# 2. Salary greater than average
print("\nEmployees earning above average salary")
cursor.execute("""
SELECT * FROM Employee
WHERE Salary >
(SELECT AVG(Salary) FROM Employee)
""")
for row in cursor.fetchall():
    print(row)

# 3. Employees working in IT using IN
print("\nEmployees in IT Department")
cursor.execute("""
SELECT Emp_Name
FROM Employee
WHERE Department IN
(SELECT Department FROM Employee WHERE Department='IT')
""")
for row in cursor.fetchall():
    print(row)

# 4. EXISTS
print("\nEmployee Exists")
cursor.execute("""
SELECT Emp_Name
FROM Employee
WHERE EXISTS
(SELECT * FROM Employee WHERE Salary>50000)
""")
for row in cursor.fetchall():
    print(row)

# 5. Create View
cursor.execute("""
CREATE VIEW Employee_View AS
SELECT Emp_Name,Department,Salary
FROM Employee
""")

# 6. Display View
print("\nEmployee View")
cursor.execute("SELECT * FROM Employee_View")
for row in cursor.fetchall():
    print(row)

# 7. Drop View
cursor.execute("DROP VIEW Employee_View")

# 8. Verify View
cursor.execute("""
SELECT name FROM sqlite_master
WHERE type='view'
AND name='Employee_View'
""")

if cursor.fetchone():
    print("View Exists")
else:
    print("\nEmployee_View deleted successfully.")

conn.close()
```

```python
#End-to-End SQL Project
import sqlite3

# Connect database
conn = sqlite3.connect("shop.db")
cursor = conn.cursor()

# Drop old objects
cursor.execute("DROP VIEW IF EXISTS Customer_Order_View")
cursor.execute("DROP TABLE IF EXISTS Orders")
cursor.execute("DROP TABLE IF EXISTS Customer")

# Create Customer table
cursor.execute("""
CREATE TABLE Customer(
    Customer_ID INTEGER PRIMARY KEY,
    Customer_Name TEXT,
    City TEXT
)
""")

# Create Orders table
cursor.execute("""
CREATE TABLE Orders(
    Order_ID INTEGER PRIMARY KEY,
    Customer_ID INTEGER,
    Order_Amount REAL,
    FOREIGN KEY(Customer_ID)
    REFERENCES Customer(Customer_ID)
)
""")

# Insert Customer records
customers=[
(1,"John","Chennai"),
(2,"Alice","Madurai"),
(3,"David","Coimbatore"),
(4,"Sophia","Chennai")
]

cursor.executemany("INSERT INTO Customer VALUES(?,?,?)",customers)

# Insert Order records
orders=[
(101,1,1200),
(102,2,2500),
(103,3,1800),
(104,1,3000),
(105,4,1500)
]

cursor.executemany("INSERT INTO Orders VALUES(?,?,?)",orders)
conn.commit()

# Display Customer table
print("Customer Table")
cursor.execute("SELECT * FROM Customer")
for row in cursor.fetchall():
    print(row)

# Display Orders table
print("\nOrders Table")
cursor.execute("SELECT * FROM Orders")
for row in cursor.fetchall():
    print(row)

# Update customer
cursor.execute("""
UPDATE Customer
SET City='Salem'
WHERE Customer_ID=2
""")
conn.commit()

# Delete an order
cursor.execute("""
DELETE FROM Orders
WHERE Order_ID=105
""")
conn.commit()

# DISTINCT
print("\nUnique Cities")
cursor.execute("SELECT DISTINCT City FROM Customer")
for row in cursor.fetchall():
    print(row)

# Aggregate functions
print("\nAggregate Functions")
cursor.execute("""
SELECT
SUM(Order_Amount),
AVG(Order_Amount),
MIN(Order_Amount),
MAX(Order_Amount)
FROM Orders
""")
print(cursor.fetchone())

# GROUP BY
print("\nOrders Grouped by Customer")
cursor.execute("""
SELECT Customer_ID,
COUNT(*),
SUM(Order_Amount)
FROM Orders
GROUP BY Customer_ID
""")
for row in cursor.fetchall():
    print(row)

# HAVING
print("\nHAVING")
cursor.execute("""
SELECT Customer_ID,
SUM(Order_Amount)
FROM Orders
GROUP BY Customer_ID
HAVING SUM(Order_Amount)>2000
""")
for row in cursor.fetchall():
    print(row)

# INNER JOIN
print("\nINNER JOIN")
cursor.execute("""
SELECT Customer.Customer_Name,
Orders.Order_Amount
FROM Customer
INNER JOIN Orders
ON Customer.Customer_ID=Orders.Customer_ID
""")
for row in cursor.fetchall():
    print(row)

# LEFT JOIN
print("\nLEFT JOIN")
cursor.execute("""
SELECT Customer.Customer_Name,
Orders.Order_Amount
FROM Customer
LEFT JOIN Orders
ON Customer.Customer_ID=Orders.Customer_ID
""")
for row in cursor.fetchall():
    print(row)

# Subquery
print("\nOrders Greater Than Average")
cursor.execute("""
SELECT *
FROM Orders
WHERE Order_Amount>
(SELECT AVG(Order_Amount)
FROM Orders)
""")
for row in cursor.fetchall():
    print(row)

# Create View
cursor.execute("""
CREATE VIEW Customer_Order_View AS
SELECT Customer.Customer_Name,
Orders.Order_Amount
FROM Customer
INNER JOIN Orders
ON Customer.Customer_ID=Orders.Customer_ID
""")

# Display View
print("\nCustomer_Order_View")
cursor.execute("SELECT * FROM Customer_Order_View")
for row in cursor.fetchall():
    print(row)

# Drop View
cursor.execute("DROP VIEW Customer_Order_View")

conn.commit()
conn.close()

print("\nView dropped successfully.")
```
